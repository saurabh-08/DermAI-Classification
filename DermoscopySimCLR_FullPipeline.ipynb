```python
# Imports
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset, WeightedRandomSampler
from torchvision import transforms, models
from PIL import Image
import os
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import CosineAnnealingLR
from tqdm import tqdm

# Warning suppression
import warnings
warnings.filterwarnings("ignore")

# Defining constants
dataset_dir = 'ISIC-images'
metadata_file = 'metadata.csv'
batch_size = 64
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Dataset Class
class DermoscopyDataset(Dataset):
    def __init__(self, csv_file, img_dir, transform=None):
        print(f"Loading dataset from {csv_file}...")
        self.data = pd.read_csv(csv_file, low_memory=False)
        print(f"CSV file loaded. Shape: {self.data.shape}")
        self.img_dir = img_dir
        self.transform = transform
        print("Dataset initialization complete.")
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, self.data.iloc[idx]['isic_id'] + '.jpg')
        image = Image.open(img_name).convert('RGB')
        label = 1 if self.data.iloc[idx]['benign_malignant'] == 'malignant' else 0
        
        if self.transform:
            image = self.transform(image)
        
        return image, label

# Model and Loss Function
class SimCLR(nn.Module):
    def __init__(self, feature_dim=128):
        super(SimCLR, self).__init__()
        self.encoder = models.resnet18(pretrained=True)
        self.encoder.fc = nn.Sequential(
            nn.Linear(self.encoder.fc.in_features, 512),
            nn.ReLU(),
            nn.Linear(512, feature_dim)
        )
        self.classifier = nn.Linear(feature_dim, 2)
    
    def forward(self, x):
        features = self.encoder(x)
        return features, self.classifier(features)

def nt_xent_loss(features, temperature=0.5):
    batch_size = features.shape[0]
    features = F.normalize(features, dim=1)
    similarity_matrix = torch.matmul(features, features.T)
    mask = torch.eye(batch_size, dtype=bool).to(device)
    positives = similarity_matrix[~mask].view(batch_size, -1)
    labels = torch.zeros(batch_size, dtype=torch.long).to(device)
    loss = F.cross_entropy(positives / temperature, labels)
    return loss

class CustomLoss(nn.Module):
    def __init__(self, weight=0.5):
        super(CustomLoss, self).__init__()
        self.ce_loss = nn.CrossEntropyLoss()
        self.weight = weight
    
    def forward(self, outputs, targets, features):
        ce_loss = self.ce_loss(outputs, targets)
        contrastive_loss = nt_xent_loss(features)
        return ce_loss + self.weight * contrastive_loss

# Utility Functions
def get_balanced_sampler(dataset):
    print("Creating balanced sampler...")
    labels = [dataset[i][1] for i in range(len(dataset))]
    class_counts = np.bincount(labels)
    class_weights = 1. / class_counts
    weights = [class_weights[label] for label in labels]
    return WeightedRandomSampler(weights, len(weights))

def stratified_split(dataset, test_size=0.2, val_size=0.2):
    print("Performing stratified split...")
    labels = [dataset[i][1] for i in range(len(dataset))]
    train_idx, test_idx = train_test_split(range(len(dataset)), test_size=test_size, stratify=labels)
    train_labels = [labels[i] for i in train_idx]
    train_idx, val_idx = train_test_split(train_idx, test_size=val_size/(1-test_size), stratify=train_labels)
    print("Stratified split complete.")
    return train_idx, val_idx, test_idx

# Training and Validation Functions
def train(model, train_loader, optimizer, criterion, epoch):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    progress_bar = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"Epoch {epoch+1}")
    for batch_idx, (inputs, targets) in progress_bar:
        inputs, targets = inputs.to(device), targets.to(device)
        
        optimizer.zero_grad()
        features, outputs = model(inputs)
        loss = criterion(outputs, targets, features)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += targets.size(0)
        correct += predicted.eq(targets).sum().item()
        
        progress_bar.set_postfix({'Loss': running_loss/(batch_idx+1), 'Acc': 100.*correct/total})
    
    return running_loss / len(train_loader), 100. * correct / total

def validate(model, val_loader, criterion):
    model.eval()
    val_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, targets in tqdm(val_loader, desc="Validating"):
            inputs, targets = inputs.to(device), targets.to(device)
            features, outputs = model(inputs)
            loss = criterion(outputs, targets, features)
            
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    
    return val_loss / len(val_loader), 100. * correct / total

# Visualization Functions
def plot_progress(train_losses, val_losses, train_accs, val_accs):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))
    ax1.plot(train_losses, label='Train Loss')
    ax1.plot(val_losses, label='Validation Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    
    ax2.plot(train_accs, label='Train Accuracy')
    ax2.plot(val_accs, label='Validation Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

def show_images(dataset, indices, titles):
    fig, axes = plt.subplots(1, len(indices), figsize=(15, 5))
    for i, idx in enumerate(indices):
        img, label = dataset[idx]
        if isinstance(img, torch.Tensor):
            img = img.permute(1, 2, 0).numpy()
        else:
            img = np.array(img)
        img = (img - img.min()) / (img.max() - img.min())
        if len(indices) == 1:
            axes.imshow(img)
            axes.set_title(f"{titles[i]} ({['Benign', 'Malignant'][label]})")
            axes.axis('off')
        else:
            axes[i].imshow(img)
            axes[i].set_title(f"{titles[i]} ({['Benign', 'Malignant'][label]})")
            axes[i].axis('off')
    plt.show()

def show_predictions(model, dataset, indices):
    model.eval()
    fig, axes = plt.subplots(1, len(indices), figsize=(15, 5))
    for i, idx in enumerate(indices):
        img, label = dataset[idx]
        img_tensor = img.unsqueeze(0).to(device)
        with torch.no_grad():
            _, output = model(img_tensor)
            _, predicted = output.max(1)
        
        img = img.permute(1, 2, 0).numpy()
        img = (img - img.min()) / (img.max() - img.min())
        if len(indices) == 1:
            axes.imshow(img)
            axes.set_title(f"True: {['Benign', 'Malignant'][label]}\nPred: {['Benign', 'Malignant'][predicted.item()]}")
            axes.axis('off')
        else:
            axes[i].imshow(img)
            axes[i].set_title(f"True: {['Benign', 'Malignant'][label]}\nPred: {['Benign', 'Malignant'][predicted.item()]}")
            axes[i].axis('off')
    plt.show()

# Main Execution
print("Defining data transforms...")
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
print("Data transforms defined.")

print("Loading datasets...")
full_dataset = DermoscopyDataset(metadata_file, dataset_dir, transform=None)
train_dataset = DermoscopyDataset(metadata_file, dataset_dir, transform=train_transform)
val_dataset = DermoscopyDataset(metadata_file, dataset_dir, transform=val_transform)
print("All datasets loaded.")

sampler = get_balanced_sampler(train_dataset)
train_idx, val_idx, test_idx = stratified_split(full_dataset)

print("Creating data loaders...")
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler, num_workers=0)
val_loader = DataLoader(Subset(val_dataset, val_idx), batch_size=batch_size, shuffle=False, num_workers=0)
test_loader = DataLoader(Subset(val_dataset, test_idx), batch_size=batch_size, shuffle=False, num_workers=0)
print("Data loaders created.")

print(f"Total samples: {len(full_dataset)}")
print(f"Training samples: {len(train_idx)}")
print(f"Validation samples: {len(val_idx)}")
print(f"Test samples: {len(test_idx)}")

# Displaying before preprocessing
benign_indices = [i for i, (_, label) in enumerate(full_dataset) if label == 0][:2]
malignant_indices = [i for i, (_, label) in enumerate(full_dataset) if label == 1][:2]
show_images(full_dataset, benign_indices + malignant_indices, ['Benign', 'Benign', 'Malignant', 'Malignant'])

# Displaying after preprocessing
show_images(train_dataset, benign_indices + malignant_indices, ['Benign (Preprocessed)', 'Benign (Preprocessed)', 'Malignant (Preprocessed)', 'Malignant (Preprocessed)'])

print("Defining SimCLR model...")
model = SimCLR().to(device)
print("SimCLR model initialized and moved to device.")

print("Setting up optimizer, scheduler, and loss function...")
optimizer = torch.optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=10)
criterion = CustomLoss()
print("Optimizer, scheduler, and loss function set up complete.")

print("Starting training loop...")
num_epochs = 20
train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    train_loss, train_acc = train(model, train_loader, optimizer, criterion, epoch)
    val_loss, val_acc = validate(model, val_loader, criterion)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    
    print(f'Train Loss: {train_loss:.3f}, Train Acc: {train_acc:.2f}%')
    print(f'Val Loss: {val_loss:.3f}, Val Acc: {val_acc:.2f}%')
    
    scheduler.step()

print("Training complete. Plotting progress...")
plot_progress(train_losses, val_losses, train_accs, val_accs)

print("Evaluating on test set...")
test_loss, test_acc = validate(model, test_loader, criterion)
print(f'Test Loss: {test_loss:.3f}, Test Acc: {test_acc:.2f}%')

print("Displaying sample predictions...")
random_indices = np.random.choice(len(val_dataset), 4, replace=False)
show_predictions(model, val_dataset, random_indices)

print("Script execution complete.")
```
